IRIS FLOWER CLASSIFICATION

The Iris flower dataset consists of three species: setosa, versicolor,
and virginica. These species can be distinguished based on their
measurements. Now, imagine that you have the measurements
of Iris flowers categorized by their respective species. Your
objective is to train a machine learning model that can learn from
these measurements and accurately classify the Iris flowers into
their respective species.

Use the Iris dataset to develop a model that can classify iris
flowers into different species based on their sepal and petal
measurements. This dataset is widely used for introductory
classification tasks.

```python
#Importing required dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Reading the Iris dataset
iris_df= pd.read_csv('/content/IRIS.csv')
```

```python
#first 5 rows of the dataset
iris_df.head()
```

```python
#last 5 rows of the dataset
iris_df.tail()
```

```python
#shape of the dataset
iris_df.shape
```

```python
#information of dataset
iris_df.info()
```

```python
#description of the dataset
iris_df.describe()
```

```python
#Checking the missing values
iris_df.isnull().sum()
```

```python
#Pie-chart for visualization of species
species = iris_df['species'].value_counts()
plt.pie(species, labels= species.index, autopct='%1.1f%%')
plt.title('Visualization of Iris Species')
```

```python
#Scatter-Plot for sepal-length and sepal-width
colors = ['red', 'orange', 'blue']
species = ['Iris-virginica','Iris-versicolor','Iris-setosa']
for i in range(3):
    x = iris_df[iris_df['species'] == species[i]]
    plt.scatter(x['sepal_length'], x['sepal_width'], c = colors[i], label=species[i])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
```

```python
#Scatter-Plot for petal-length and petal-width
for i in range(3):
    x = iris_df[iris_df['species'] == species[i]]
    plt.scatter(x['petal_length'], x['petal_width'], c = colors[i], label=species[i])
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.legend()
```

```python
#Correlation Plot
corr = iris_df.corr(numeric_only=True)
fig, ax = plt.subplots(figsize=(5,4))
sns.heatmap(corr, annot=True, ax=ax, cmap = 'coolwarm')
```

```python
#Splitting into data and target
X=iris_df[['sepal_length','sepal_width', 'petal_length', 'petal_width']]
Y=iris_df['species']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30)
```

```python
# loading the logistic regression model
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
print("Accuracy: ",model.score(X_test, Y_test) * 100)
```

```python
#loading the KNeighbors Classifier model
knc_model = KNeighborsClassifier()
```

```python
knc_model.fit(X_train, Y_train)
```

```python
print("Accuracy: ",knc_model.score(X_test, Y_test) * 100)
```
