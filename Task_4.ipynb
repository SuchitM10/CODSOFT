SALES PREDICTION USING PYTHON

Sales prediction involves forecasting the amount of a product that
customers will purchase, taking into account various factors such as
advertising expenditure, target audience segmentation, and
advertising platform selection.

In businesses that offer products or services, the role of a Data
Scientist is crucial for predicting future sales. They utilize machine
learning techniques in Python to analyze and interpret data, allowing
them to make informed decisions regarding advertising costs. By
leveraging these predictions, businesses can optimize their
advertising strategies and maximize sales potential. Let's embark on
the journey of sales prediction using machine learning in Python.

```python
#Importing required dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
#creating dataframe using car_purchasing datasets
car_data_df = pd.read_csv('/content/car_purchasing.csv',encoding='ISO-8859-1')
```

```python
#first 5 rows of the dataset
car_data_df.head()
```

```python
#last 5 rows of dataset
car_data_df.tail()
```

```python
#checking shape of dataset
car_data_df.shape
```

```python
#dataset information
car_data_df.info()
```

```python
#description of dataset
car_data_df.describe()
```

```python
# checking missing values in each column
car_data_df.isnull().sum()
```

```python
cols = ['country','gender', 'age',
       'annual Salary', 'credit card debt', 'net worth']
plt.figure(figsize=(18, 12))
for i in range(6) :
    plt.subplot(2,3,i+1)
    plt.scatter(car_data_df[cols[i]],car_data_df['car purchase amount'])
    plt.title(cols[i])
    plt.ylabel('car purchase amount',size=13)
plt.show()
```

```python
#Splitting into data and target
X=car_data_df[['gender', 'age',
       'annual Salary', 'credit card debt', 'net worth']]
Y=car_data_df['car purchase amount']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting training and testing data
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2,shuffle=True)
```

```python
# loading the linear regression model
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
# prediction on Training data
training_data_prediction = model.predict(X_train)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
#training data visualization
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
# prediction on testing data
test_data_prediction = model.predict(X_test)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared Error : ", error_score)
```

```python
#testing data visualization
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```
