CREDIT CARD FRAUD DETECTION

Build a machine learning model to identify fraudulent credit card transactions.

Preprocess and normalize the transaction data, handle class imbalance issues, and split the dataset into training and testing sets.

Train a classification algorithm, such as logistic regression or random forests, to classify transactions as fraudulent or genuine.

Evaluate the model's performance using metrics like precision, recall, and F1-score, and consider techniques like oversampling or undersampling for improving results.

```python
#Importing necessary dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
```

```python
#loading dataset
credit_card_data = pd.read_csv("/content/creditcard.csv")
```

```python
# first 5 rows of the dataset
credit_card_data.head()
```

```python
# last 5 rows of the dataset
credit_card_data.tail()
```

```python
# dataset informations
credit_card_data.info()
```

```python
# checking missing values in each column
credit_card_data.isnull().sum()
```

```python
#filling the missing values with null strings
credit_card_data = credit_card_data.fillna('')
```

```python
# checking distribution of legitimate transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

We have a highly imbalanced dataset.

0 -> Legitimate Transactions

1 -> Fraudulent Transactions

```python
# preparing the dataset for analysis
legitimate = credit_card_data[credit_card_data.Class == 0]
fraudulent = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legitimate.shape)
print(fraudulent.shape)
```

```python
#statistical description of dataset
legitimate.Amount.describe()
```

```python
fraudulent.Amount.describe()
```

```python
# comparing the values for both transactions
credit_card_data.groupby('Class').mean()
```

Under-Sampling

```python
#creating sample dataset containing similar distribution of legitimate transactions and Fraudulent Transactions
legitimate_sample = legitimate.sample(n=81)
```

```python
#concatenating two datasets
concat_dataset = pd.concat([legitimate_sample, fraudulent], axis=0)
```

```python
concat_dataset.head()
```

```python
concat_dataset.tail()
```

```python
concat_dataset.groupby('Class').mean()
```

```python
#splitting the dataset into features and columns
X = concat_dataset.drop(columns='Class', axis=1)
Y = concat_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Logistic Regression Model

```python
model = LogisticRegression(solver='lbfgs', max_iter=2000)
```

```python
# training the Logistic Regression Model using training Data
model.fit(X_train,Y_train)
```

```python
# precision on training data
X_train_prediction = model.predict(X_train)
training_data_precision = precision_score(X_train_prediction, Y_train)
training_data_f1_score = f1_score(X_train_prediction, Y_train)
training_data_recall_score = recall_score(X_train_prediction, Y_train)
```

```python
print('Precision on Training data : ', training_data_precision)
print('F1-score on training data: ', training_data_f1_score)
print('Recall on training data: ', training_data_recall_score)
```

```python
# precision on test data
X_test_prediction = model.predict(X_test)
testing_data_precision = precision_score(X_test_prediction, Y_test)
testing_data_f1_score = f1_score(X_test_prediction, Y_test)
testing_data_recall_score = recall_score(X_test_prediction, Y_test)
```

```python
print('Precision on testing data : ', testing_data_precision)
print('F1-score on testing data: ', testing_data_f1_score)
print('Recall on testing data: ', testing_data_recall_score)
```
